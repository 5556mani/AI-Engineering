```python
import torch.nn as nn
import torch
```

```python
#
```

```python
class One_hidden_NN(nn.Module):

  def __init__(self,input_shape):

    super().__init__()
    self.linear=nn.Linear(input_shape,3)
    self.relu=nn.ReLU()
    self.hidden=nn.Linear(3,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self, features):

    out=self.linear(features)
    out=self.relu(out)
    out=self.hidden(out)
    out=self.sigmoid(out)

    return out
```

```python
dummy=torch.rand((5,7))
```

```python
model=One_hidden_NN(dummy.shape[1])
```

```python
class One_hidden_NN(nn.Module):

  def __init__(self,input_shape):

    super().__init__()
    self.seq=nn.Sequential(
                              nn.Linear(input_shape,3),
                              nn.ReLU(),
                              nn.Linear(3,1),
                              nn.Sigmoid()
                                                         )

  def forward(self, features):

    out=self.seq(features)

    return out
```
