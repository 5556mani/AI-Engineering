```python
import torch

# Start with a simple 1D tensor (like a basic list)
x = torch.tensor([5, 6, 7])
print("Original x:")
print(x)
print("Shape:", x.shape)  # Output: torch.Size([3])
print("-" * 30)

# ==========================================
# 1. UNSQUEEZE: Adding a dimension
# ==========================================

# Let's add a dimension at index 0 (the very front)
# This turns our 1D array into a 2D matrix with 1 row and 3 columns.
x_unsqueezed_0 = x.unsqueeze(dim=0)
print("\nUnsqueeze at dim=0:")
print(x_unsqueezed_0)
print("Shape:", x_unsqueezed_0.shape)  # Output: torch.Size([1, 3])

# What if we add a dimension at index 1 instead?
# This turns it into a 2D matrix with 3 rows and 1 column (a column vector).
x_unsqueezed_1 = x.unsqueeze(dim=1)
print("\nUnsqueeze at dim=1:")
print(x_unsqueezed_1)
print("Shape:", x_unsqueezed_1.shape)  # Output: torch.Size([3, 1])
print("-" * 30)

# ==========================================
# 2. SQUEEZE: Removing a dimension
# ==========================================

# Let's take that [1, 3] tensor and squish it back down!
# If you don't pass a dimension, .squeeze() looks for ANY dimension
# of size 1 and removes it.
x_squeezed = x_unsqueezed_0.squeeze()
print("\nSqueeze (removes the '1'):")
print(x_squeezed)
print("Shape:", x_squeezed.shape)  # Output: torch.Size([3])

# Note: .squeeze() ONLY removes dimensions that are exactly size 1.
# If you try to squeeze a tensor with shape [2, 3], nothing happens!
```
