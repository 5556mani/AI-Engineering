```python
# clearing grad
# Z.backward() directly running this will throw error you have to make one forward pass as old computation graph is destroyed now
import torch
X=torch.tensor([1,2,3])
X=X.to(dtype=torch.float32)
X.requires_grad_(True)
```

```python
X.grad.zero_()
Z=(X**2).mean(dtype=float)
Z.backward()
X.grad


# if you run it multiple time values start geeting add
# to prevent this you need to call .grad.zero_()
```

```python
# to stop gradiect tracking
# --requires_grad_ , detach , no_grad
```

```python
import torch

X = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)

# Toggle the tracking flag to False in-place
X.requires_grad_(False)

# Any operation downstream will NOT have a computation graph
Z = (X**2).mean()

print(Z.requires_grad)  # Output: False
# Z.backward()          # This will CRASH now because there is no graph
```

```python
import torch

X = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)

# Create a severed view
X_detached = X.detach()

# The original X still tracks gradients, but X_detached does not
Y1 = (X**2).mean()          # Tracks gradients
Y2 = (X_detached**2).mean() # Does NOT track gradients

print("Original tracking:", Y1.requires_grad)  # Output: True
print("Detached tracking:", Y2.requires_grad)  # Output: False
```

```python
import torch

X = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)

# Everything inside this block has tracking suspended
with torch.no_grad():
    Y = X ** 2
    Z = Y.mean()
    print("Inside block tracking:", Z.requires_grad)  # Output: False

# As soon as you unindent, tracking turns back on automatically
Out = X * 2
print("Outside block tracking:", Out.requires_grad)  # Output: True
```
