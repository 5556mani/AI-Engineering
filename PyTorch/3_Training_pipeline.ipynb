```python
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import StandardScaler as ss
from sklearn.preprocessing import LabelEncoder as le
from sklearn.model_selection import train_test_split as tts
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
```

```python
X_train , X_test , Y_train , Y_test = tts(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
ss=ss()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
le=le()
Y_train= le.fit_transform(Y_train)
Y_test=le.transform(Y_test)
```

```python
X_train=torch.from_numpy(X_train)
X_test=torch.from_numpy(X_test)
Y_train=torch.from_numpy(Y_train)
Y_test=torch.from_numpy(Y_test)
```

# Define the Model using class


1.   def Class
2.   def init
3.   def forward_pass
4.   def loss function



```python
class Model():

  def __init__(self,X):
    self.weights=torch.rand((X.shape[1],1),dtype=torch.float64,requires_grad=True)
    self.bias=torch.tensor([1],dtype=torch.float64,requires_grad=True)

  def forward_pass(self,X):
    X=torch.matmul(X,self.weights)
    X=X+self.bias
    return X

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(Y_train * torch.log(y_pred) + (1 - Y_train) * torch.log(1 - y_pred)).mean()
    return loss
```

# Input Parameters


```python
learning_rate=0.1
epoch=50
```

# Training Pipeline

```python
model=Model(X_train)

print(model.weights.shape)
print(X_train.shape)

for epoch in range(epoch):
  Y_pred=model.forward_pass(X_train)

  loss=model.loss_function(Y_pred,Y_train)

  loss.backward()

  with torch.no_grad():
    model.weights-=model.weights.grad * learning_rate
    model.bias-=model.bias.grad * learning_rate

  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f"epoch: {epoch}  loss: {loss}")
```
